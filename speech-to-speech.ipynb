```python
import torch                        # allows Tensor computation with strong GPU acceleration
from transformers import pipeline   # fast way to use pre-trained models for inference
import os
```

```python
# load model
dolly_pipeline = pipeline(model="databricks/dolly-v2-3b",
                            torch_dtype=torch.bfloat16,
                            trust_remote_code=True,
                            device_map="auto")
```

```python
# define helper function
def generate_response(input):
  system = f""" Welcome to Dolly, the new generation language model from Databricks."""

  prompt = f"""
  System: {system}\n
  User: {input}\n
  Response from Dolly-v2-3b:
  """
  print(prompt)
  dolly_response = dolly_pipeline(prompt,
                                  max_new_tokens=500
                                  )
  return dolly_response[0]["generated_text"]
```

### Util function to return text file from audio file

```python
# open the file
import speech_recognition as sr

def get_text(filename):
  r = sr.Recognizer()

  with sr.AudioFile(filename) as source:
      # listen for the data (load audio to memory)
      audio_data = r.record(source)
      # recognize (convert from speech to text)
      text = r.recognize_google(audio_data)
      return text
```

```python
from gtts import gTTS
from IPython.display import Audio, display

def text_to_speech(text):
    tts = gTTS(text=text, lang='en')
    tts.save('output.mp3')
    display(Audio('output.mp3', autoplay=True))

# step - 1 converting audio to text format
filename = "/content/record_out_1.wav"
prompt = get_text(filename)

# step - 2 sending prompt to generate response
response = generate_response(prompt)
print(response,'\n\n') # print response

# step - 3 convert back generated response to speech
text_to_speech(response)
```
